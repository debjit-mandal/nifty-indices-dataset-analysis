# Sector-wise Performance Analysis of NIFTY Indices
This notebook compares the performance of different sectors in NIFTY indices to identify sector leaders and laggards.


```python
import pandas as pd
import matplotlib.pyplot as plt

nifty_auto_df = pd.read_csv('data/NIFTY AUTO.csv')
nifty_it_df = pd.read_csv('data/NIFTY IT.csv')
nifty_fmcg_df = pd.read_csv('data/NIFTY FMCG.csv')
nifty_bank_df = pd.read_csv('data/NIFTY BANK.csv')
nifty_metal_df = pd.read_csv('data/NIFTY METAL.csv')
nifty_pharma_df = pd.read_csv('data/NIFTY PHARMA.csv')

sector_dataframes = {
    "NIFTY AUTO": nifty_auto_df,
    "NIFTY IT": nifty_it_df,
    "NIFTY FMCG": nifty_fmcg_df,
    "NIFTY BANK": nifty_bank_df,
    "NIFTY METAL": nifty_metal_df,
    "NIFTY PHARMA": nifty_pharma_df
}

for df in sector_dataframes.values():
    df['Date'] = pd.to_datetime(df['Date'])

# Plotting sector-specific indices
plt.figure(figsize=(14, 8))
for name, df in sector_dataframes.items():
    if 'Close' in df.columns:
        plt.plot(df['Date'], df['Close'], label=name)

plt.title('Closing Prices Over Time for Various NIFTY Sector Indices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

Comparison of closing prices across sectors

```python

# Comparing closing prices across sectors
plt.figure(figsize=(14, 8))
for name, df in sector_dataframes.items():
    if 'Close' in df.columns:
        plt.plot(df['Date'], df['Close'], label=name)

plt.title('Comparison of Closing Prices Across Sectors')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

Sector-wise volume analysis

```python
# Sector-wise volume analysis
plt.figure(figsize=(14, 8))
for name, df in sector_dataframes.items():
    if 'Volume' in df.columns:
        plt.plot(df['Date'], df['Volume'], label=name)

plt.title('Sector-wise Volume Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

Sector-wise turnover analysis

```python

# Sector-wise turnover analysis
plt.figure(figsize=(14, 8))
for name, df in sector_dataframes.items():
    if 'Turnover' in df.columns:
        plt.plot(df['Date'], df['Turnover'], label=name)

plt.title('Sector-wise Turnover Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Turnover')
plt.legend()
plt.grid(True)
plt.show()
```

Sector leaders and laggards

```python
# Identifying sector leaders and laggards
cumulative_returns = {}

for name, df in sector_dataframes.items():
    df.set_index('Date', inplace=True)
    cumulative_returns[name] = (df['Close'] / df['Close'].iloc[0]) - 1

# Plotting cumulative returns
plt.figure(figsize=(14, 8))
for name, returns in cumulative_returns.items():
    plt.plot(returns, label=name)

plt.title('Cumulative Returns for Various Sectors')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()

```

Final cumulative returns to identify leaders and laggards

```python
final_returns = {name: returns.iloc[-1] for name, returns in cumulative_returns.items()}
final_returns_df = pd.DataFrame(list(final_returns.items()), columns=['Sector', 'Cumulative Return'])
final_returns_df.sort_values(by='Cumulative Return', ascending=False, inplace=True)
final_returns_df
```
