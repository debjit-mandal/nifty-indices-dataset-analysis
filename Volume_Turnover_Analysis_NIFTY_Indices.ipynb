# Volume and Turnover Analysis of NIFTY Indices
This notebook performs an analysis of volume and turnover trends to understand market activity in NIFTY indices.


```python
import pandas as pd
import matplotlib.pyplot as plt

nifty_50_df = pd.read_csv('data/NIFTY 50.csv')
nifty_100_df = pd.read_csv('data/NIFTY 100.csv')
nifty_500_df = pd.read_csv('data/NIFTY 500.csv')
nifty_auto_df = pd.read_csv('data/NIFTY AUTO.csv')
nifty_bank_df = pd.read_csv('data/NIFTY BANK.csv')
nifty_fmcg_df = pd.read_csv('data/NIFTY FMCG.csv')
nifty_it_df = pd.read_csv('data/NIFTY IT.csv')
nifty_metal_df = pd.read_csv('data/NIFTY METAL.csv')
nifty_microcap_250_df = pd.read_csv('data/NIFTY MICROCAP 250.csv')
indiavix_df = pd.read_csv('data/INDIAVIX.csv')
nifty_midcap_150_df = pd.read_csv('data/NIFTY MIDCAP 150.csv')
nifty_next_50_df = pd.read_csv('data/NIFTY NEXT 50.csv')
nifty_pharma_df = pd.read_csv('data/NIFTY PHARMA.csv')
nifty_smallcap_250_df = pd.read_csv('data/NIFTY SMALLCAP 250.csv')

dataframes = {
    "NIFTY 50": nifty_50_df,
    "NIFTY 100": nifty_100_df,
    "NIFTY 500": nifty_500_df,
    "NIFTY AUTO": nifty_auto_df,
    "NIFTY BANK": nifty_bank_df,
    "NIFTY FMCG": nifty_fmcg_df,
    "NIFTY IT": nifty_it_df,
    "NIFTY METAL": nifty_metal_df,
    "NIFTY MICROCAP 250": nifty_microcap_250_df,
    "INDIAVIX": indiavix_df,
    "NIFTY MIDCAP 150": nifty_midcap_150_df,
    "NIFTY NEXT 50": nifty_next_50_df,
    "NIFTY PHARMA": nifty_pharma_df,
    "NIFTY SMALLCAP 250": nifty_smallcap_250_df
}

for df in dataframes.values():
    df['Date'] = pd.to_datetime(df['Date'])

# Plotting volume trends over time
plt.figure(figsize=(14, 8))
for name, df in dataframes.items():
    if 'Volume' in df.columns:
        plt.plot(df['Date'], df['Volume'], label=name)

plt.title('Volume Trends Over Time for Various NIFTY Indices')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

Turnover values

```python
# Turnover values
plt.figure(figsize=(14, 8))
for name, df in dataframes.items():
    if 'Turnover' in df.columns:
        plt.plot(df['Date'], df['Turnover'], label=name)

plt.title('Turnover Values Over Time for Various NIFTY Indices')
plt.xlabel('Date')
plt.ylabel('Turnover')
plt.legend()
plt.grid(True)
plt.show()
```

High and low activity periods for NIFTY 50

```python
nifty_50_df.set_index('Date', inplace=True)

plt.figure(figsize=(14, 8))
has_data = False
if 'Volume' in nifty_50_df.columns:
    plt.plot(nifty_50_df.index, nifty_50_df['Volume'], label='Volume')
    has_data = True
if 'Turnover' in nifty_50_df.columns:
    plt.plot(nifty_50_df.index, nifty_50_df['Turnover'], label='Turnover')
    has_data = True

if has_data:
    plt.title('NIFTY 50 Volume and Turnover Over Time')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("No Volume or Turnover data available for NIFTY 50")
```

Correlation between volume, turnover, and closing prices for NIFTY 50

```python
# Correlation between volume, turnover, and closing prices for NIFTY 50
columns_to_corr = ['Close']
if 'Volume' in nifty_50_df.columns:
    columns_to_corr.append('Volume')
if 'Turnover' in nifty_50_df.columns:
    columns_to_corr.append('Turnover')

if len(columns_to_corr) > 1:
    nifty_50_corr = nifty_50_df[columns_to_corr].corr()
    
    plt.figure(figsize=(8, 6))
    plt.imshow(nifty_50_corr, cmap='coolwarm', interpolation='none')
    plt.colorbar()
    plt.xticks(range(len(nifty_50_corr)), nifty_50_corr.columns, rotation=90)
    plt.yticks(range(len(nifty_50_corr)), nifty_50_corr.columns)
    plt.title('Correlation Matrix for NIFTY 50')
    plt.show()
else:
    print("Not enough data to compute correlations for NIFTY 50")
```
